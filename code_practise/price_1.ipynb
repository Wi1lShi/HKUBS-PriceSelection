```python
import matplotlib.pyplot as plt

# Initialize lists to store prices and profits
prices = []
profits = []

# Open and read the file
with open('profit.txt', 'r') as file:
    for line in file:
        # Split the line by commas and extract the price and total profit
        parts = line.split(',')
        price = float(parts[0].split(':')[1].strip())
        profit = float(parts[3].split(':')[1].strip())
        
        # Append to the lists
        prices.append(price)
        profits.append(profit)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(prices, profits, marker='o')
plt.title('Price vs Total Profit')
plt.xlabel('Price')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```
